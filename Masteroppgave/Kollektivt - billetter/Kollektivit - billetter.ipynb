```python
import requests
import pandas as pd
from pyjstat import pyjstat

# URL til SSB API
url = "https://data.ssb.no/api/v0/no/table/11570/"

# Spørringen i JSON-format
query = {
    "query": [
        {
            "code": "TransportA",
            "selection": {
                "filter": "item",
                "values": [
                    "00",
                    "10",
                    "20",
                    "32",
                    "31"
                ]
            }
        },
        {
            "code": "Ruteform",
            "selection": {
                "filter": "item",
                "values": [
                    "00",
                    "01",
                    "02"
                ]
            }
        }
    ],
    "response": {
        "format": "json-stat2"
    }
}

# Sender forespørselen
response = requests.post(url, json=query)

# Sjekk status for responsen
if response.status_code == 200:
    print("Data hentet vellykket!")
    data = response.json()
else:
    print(f"Feil ved henting av data: {response.status_code}")
    data = None

# Konverterer JSON-stat data til Pandas DataFrame
if data:
    # Bruker pyjstat for å konvertere JSON-stat til DataFrame
    dataset = pyjstat.Dataset(data)
    df = dataset.write('dataframe')
    
    # Skriv ut de første radene i DataFrame
    print(df.head())
    
    # Lagre til en CSV-fil (valgfritt)
    df.to_csv("ssb_data.csv", index=False)
    print("Data lagret som 'ssb_data.csv'")
```

```python
import requests
import pandas as pd
from pyjstat import pyjstat

# URL til SSB API
url = "https://data.ssb.no/api/v0/no/table/11570/"

# Spørringen i JSON-format
query = {
    "query": [
        {
            "code": "TransportA",
            "selection": {
                "filter": "item",
                "values": [
                    "00",
                    "10",
                    "20",
                    "32",
                    "31"
                ]
            }
        },
        {
            "code": "Ruteform",
            "selection": {
                "filter": "item",
                "values": [
                    "00",
                    "01",
                    "02"
                ]
            }
        }
    ],
    "response": {
        "format": "json-stat2"
    }
}

# Sender forespørselen
response = requests.post(url, json=query)

# Sjekk status for responsen
if response.status_code == 200:
    print("Data hentet vellykket!")
    data = response.json()
else:
    print(f"Feil ved henting av data: {response.status_code}")
    data = None

# Konverter JSON-stat data til Pandas DataFrame
if data:
    # Bruker pyjstat for å konvertere JSON-stat til DataFrame
    dataset = pyjstat.Dataset(data)
    df = dataset.write('dataframe')
    
    # Skriv ut de første radene i DataFrame
    print(df.head())
    
    # Lag en pivot-tabell med "statistikkvariabel" og "transportmåte" som kolonner
    pivot_df = df.pivot_table(
        index='år',  # År som rader
        columns=['statistikkvariabel', 'transportmåte'],  # Statistikkvariabel og transportmåte som kolonner
        values='value',  # Verdier fra 'value'-kolonnen
        aggfunc='sum'  # Summer verdier hvis det finnes duplikater
    )
    
    # Fjern multiindeks i kolonner (valgfritt, hvis du ønsker flatere struktur)
    pivot_df.columns = ['_'.join(col).strip() for col in pivot_df.columns.values]
    
    # Skriv ut de første radene av pivot-tabellen
    print(pivot_df.head())
    
    # Lagre pivot-tabellen til en CSV-fil
    pivot_df.to_csv("ssb_data_transport_statistikkvariabel.csv")
    print("Pivotert data lagret som 'ssb_data_transport_statistikkvariabel.csv'")
```
